# Impulse Response

```python
import numpy as np
import matplotlib.pyplot as plt

# Impulse response h[n]
h = np.array([0.2, 0.5, 0.3])

# Input signal x[n]
x = np.array([1, 2, 3, 4, 5])

# Impulse signal
impulse = np.zeros_like(x)
impulse[0] = 1

# Calculate the output y[n] by convolving x[n] with h[n]
y = np.convolve(x, h)

# Plot the input signal x[n]
plt.figure(figsize=(12, 10))
plt.subplot(3, 1, 1)
plt.stem(x)
plt.grid()
plt.title('Input Signal x[n]')
plt.xlabel('n')
plt.ylabel('Amplitude')

# Plot the impulse response h[n]
plt.subplot(3, 1, 2)
plt.stem(h)
plt.grid()
plt.title('Impulse Response h[n]')
plt.xlabel('n')
plt.ylabel('Amplitude')

# Plot the output signal y[n]
plt.subplot(3, 1, 3)
plt.stem(y)
plt.title('Output Signal y[n] (Convolution of x[n] and h[n])')
plt.xlabel('n')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.grid()
plt.savefig("Impulse Response.png")
plt.show()
```

# Frequency Response 

```python
# Define the impulse response (example FIR filter coefficients)
h = np.array([0.2, 0.5, 0.3])

# Compute the frequency response using FFT (Fast Fourier Transform)
H = np.fft.fft(h, 1024)  # FFT size 1024 for better frequency resolution

# Sampling frequency (Hz)
f_s = 1000  # Example value, adjust as needed

# Frequency axis in Hz
frequencies = np.linspace(0, f_s / 2, 1024 // 2)  # Half the FFT points due to symmetry

# Magnitude and phase responses
magnitude_response = 20 * np.log10(np.abs(H[:1024 // 2]))  # Convert to dB
phase_response = np.angle(H[:1024 // 2])

# Plotting
plt.figure(figsize=(10, 6))

# Magnitude Response Plot
plt.subplot(2, 1, 1)
plt.plot(frequencies, magnitude_response, label=f'Impulse Response: {h}, Sampling Rate: {f_s} Hz')
plt.grid(which='both', axis='both')
plt.title('Magnitude Response')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude (dB)')
plt.legend()

# Phase Response Plot
plt.subplot(2, 1, 2)
plt.plot(frequencies, phase_response)
plt.title('Phase Response')
plt.grid(which='both', axis='both')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Phase (radians)')

plt.tight_layout()
plt.savefig("Frequency Response.png")
plt.show()
```
